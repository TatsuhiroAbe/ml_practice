```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
df['Legendary'] = df['Legendary'].astype(int)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12 , 8))
df.plot(kind='scatter', x='HP', y='Attack', s=100, 
        c='Legendary', cmap='winter', alpha=0.5, ax=ax)
```

```python
df.loc[:, 'HP':'Speed'].corr()
```

```python
sns.pairplot(df.loc[:, 'HP':'Speed'])
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X_pca = pca.fit_transform(df.loc[:, 'HP':'Speed'])
df_pca = pd.DataFrame(X_pca, columns=['1st', '2nd', '3rd', '4th', '5th', '6th'])
df_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10, 8))
plt.plot(np.hstack([0, pca.explained_variance_ratio_.cumsum()]))
plt.xlabel("n_components", fontsize=15)
plt.ylabel("explained_variance_ratio_", fontsize=15)
plt.show()
```

```python
pca.components_
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(pca.components_, 
            cmap='Blues', 
            annot=True, 
            annot_kws={"size": 20},
            fmt="1.1f",
            xticklabels=df.loc[:, 'HP':'Speed'].columns,
            yticklabels=['1st', '2nd', '3rd', '4th', '5th', '6th'])
```

```python
df_pca['Legendary'] = df['Legendary']

fig, ax = plt.subplots(figsize=(12 , 8))
df_pca.plot(kind='scatter', x='1st', y='2nd', s=100, 
            c='Legendary', cmap='winter', alpha=0.5, ax=ax)
```

```python
from mpl_toolkits.mplot3d import Axes3D

y = df['Legendary']

fig, ax = plt.subplots(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
    
ax.scatter(X_pca[:, 0][y==0], X_pca[:, 1][y==0], color='#0b00fb', s=100, alpha=0.5)
ax.scatter(X_pca[:, 0][y==1], X_pca[:, 1][y==1], color='#90ff88', s=100, alpha=0.5)

ax.set_xlabel("First Principal Component", fontsize=15)
ax.set_ylabel("Second Principal Component", fontsize=15)
ax.set_zlabel("Third Principal Component", fontsize=15)

ax.legend(['Not Legendary', 'Legendary'], loc="best", fontsize=16)
    
plt.show()
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
print('before: ', df.shape)
df = df.drop(columns=['LotFrontage', 'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'], axis=1)
print('after: ', df.shape)
```

```python
df.dtypes.value_counts()
```

```python
all_columns = df.columns.values
non_cat_cols = ["LotFrontage", "LotArea", "MasVnrArea", "BsmtFinSF1", 
                "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF", "1stFlrSF", 
                "2ndFlrSF", "LowQualFinSF", "GrLivArea", "GarageArea", 
                "WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", 
                "ScreenPorch","PoolArea", "MiscVal"]

cat_cols = [col for col in all_columns if col not in non_cat_cols]
```

```python
nan_cols = df.isnull().sum()[df.isnull().sum() > 0].index

for col in nan_cols:
    if col in non_cat_cols:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
X = df.drop(columns=['Id', 'SalePrice'])
y = df['SalePrice']

X = np.log1p(X)
y = np.log1p(y)

X[X == -np.inf] = 0

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
pca = PCA()
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)

np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(np.hstack([0, pca.explained_variance_ratio_.cumsum()]))
plt.xlabel("n_components", fontsize=15)
plt.ylabel("explained_variance_ratio_", fontsize=15)
plt.show()
```

```python
X_train_pca = X_train_pca[:, :8]
X_test_pca = X_test_pca[:, :8]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_absolute_error
```

```python
%%time

cv = KFold(n_splits=5, shuffle=True, random_state=0)

rf_scores = np.sqrt(-cross_val_score(RandomForestRegressor(n_estimators=500, max_depth=6),
                             X=X_train,
                             y=y_train,
                             cv=cv,
                             scoring='neg_mean_squared_error'))
print(rf_scores.mean())
```

```python
%%time

cv = KFold(n_splits=5, shuffle=True, random_state=0)

rf_pca_scores = np.sqrt(-cross_val_score(RandomForestRegressor(n_estimators=500, max_depth=6),
                                 X=X_train_pca,
                                 y=y_train,
                                 cv=cv,
                                 scoring='neg_mean_squared_error'))
print(rf_pca_scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
%%time

rf_param_grid = {
    'max_depth': range(5, 11)
}

rf_scores = np.sqrt(-cross_val_score(GridSearchCV(RandomForestRegressor(n_estimators=500), 
                                         param_grid=rf_param_grid, 
                                         cv=cv),
                            X=X_train, 
                            y=y_train, 
                            cv=cv, 
                            scoring='neg_mean_absolute_error'))

print(rf_scores.mean())
```

```python
%%time

rf_param_grid = {
    'max_depth': range(5, 11)
}

rf_pca_scores = np.sqrt(-cross_val_score(GridSearchCV(RandomForestRegressor(n_estimators=500), 
                                         param_grid=rf_param_grid, 
                                         cv=cv),
                            X=X_train_pca, 
                            y=y_train, 
                            cv=cv, 
                            scoring='neg_mean_absolute_error'))

print(rf_pca_scores.mean())
```
